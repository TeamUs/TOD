````python
import numpy as np
import matplotlib.pyplot as plt

lg = {
    0: "waffle iron french toast",
    1: "zwetschgenkuchen bavarian plum cake",
    2: "lime tea"
}


# загрузка данных
data = np.load('average_ratings.npy')

# создание фигуры и осей
fig, ax = plt.subplots()

# настройка осей
ax.set_xlabel('Номер дня')
ax.set_ylabel('Средний рейтинг')
ax.set_title('Изменение среднего рейтинга трех рецептов')

# построение временных рядов
for i in range(3):
    ax.plot(data[i], label=f'{lg[i]}')

# добавление легенды
ax.legend()

# отображение графика
plt.show()
````

Далее настройте `major_locator` и `minor_locator` горизонтальной оси (подробнее см. [здесь](https://matplotlib.org/stable/gallery/text_labels_and_annotations/date.html))

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

ratings = np.load('average_ratings.npy')

dates = pd.date_range('2019-01-01', '2021-12-30', freq='D')
years = mdates.YearLocator()   # засечки для годов
months = mdates.MonthLocator()  # засечки для месяцев
years_fmt = mdates.DateFormatter('%Y')  # формат подписей для годов

fig, ax = plt.subplots(figsize=(10, 5))

for i, recipe in enumerate(ratings):
    ax.plot(dates, recipe, label=f"{lg[i]}")
    
ax.set_title("Изменение среднего рейтинга трех рецептов")
ax.set_ylabel("Средний рейтниг")
ax.legend()

ax.xaxis.set_major_locator(years)
ax.xaxis.set_minor_locator(months)
ax.xaxis.set_major_formatter(years_fmt)

ax.set_xlabel("Дата")

plt.show()
````

5\. Назовем рецепты короткими, если они занимают строго меньше 5 минут; средними, если они занимают от 5 до 50 минут (не включая 50), и длинными, если они занимают от 50 минут и больше. Сгруппируйте все рецепты по данному признаку и для каждой группы посчитайте 2 величины: среднее количество шагов рецептов в группе и размер группы. При помощи методов структур `pandas` постройте столбчатую диаграмму, где каждый столбец означает группу (короткие, средние или длинные рецепты), а высота столбца обозначает среднее количество шагов рецептов в группе. Рядом по горизонтали разместите круговую диаграмму, на которой отображены размеры каждой из групп.

Добавьте следующие подписи:
* по горизонтальной оси под столбчатой диаграммой напишите "Группа рецептов"
* по вертикальной оси слева от столбчатой диаграммы напишите "Средняя длительность"
* над круговой диаграммой напишите "Размеры групп рецептов"

````python
import pandas as pd
import matplotlib.pyplot as plt

reviews = pd.read_csv('reviews_sample.csv')
recipes = pd.read_csv('recipes_sample.csv')

# Группировка рецептов по длительности приготовления
short_recipes = recipes[recipes['minutes'] < 5]
medium_recipes = recipes[(recipes['minutes'] >= 5) & (recipes['minutes'] < 50)]
long_recipes = recipes[recipes['minutes'] >= 50]

# Подсчет среднего количества шагов и размера группы для каждой группы рецептов
data = {'Средняя длительность': [short_recipes['n_steps'].mean(), medium_recipes['n_steps'].mean(), long_recipes['n_steps'].mean()],
        'Размеры групп рецептов': [len(short_recipes), len(medium_recipes), len(long_recipes)]}
df = pd.DataFrame(data=data, index=['Короткие', 'Средние', 'Длинные'])

# Создание холста с двумя графиками
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15, 8))

# Построение столбчатой диаграммы
df['Средняя длительность'].plot(kind='bar', color='blue', alpha=0.5, width=0.5, ax=ax1)
ax1.set_xlabel('Группа рецептов')
ax1.set_ylabel('Средняя длительность')
ax1.set_title('Средняя длительность рецептов по группам')

# Добавление значений над столбцами
for p in ax1.patches:
    ax1.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

# Построение круговой диаграммы
df['Размеры групп рецептов'].plot(kind='pie', colors=['orange', 'green', 'red'], autopct='%1.1f%%', explode=(0, 0.05, 0.1), startangle=90, radius=0.9, ax=ax2)
ax2.set_title('Размеры групп рецептов')

# Отображение графиков
plt.show()
````

8\. Объедините две таблицы `recipes` и `reviews` и постройте корреляционную матрицу на основе столбцов "minutes", "n_steps", "n_ingredients" и "rating". При помощи пакета `seaborn` визуализируйте полученную матрицу в виде тепловой карты (heatmap). 

Добавьте в ячейки тепловой карты подписи (значения к-та корреляции). Измените цветовую палитру на `YlOrRd`. 

Добавьте заголовок рисунка "Корреляционная матрица числовых столбцов таблиц recipes и reviews"

````python
import pandas as pd
import seaborn as sns

# Загрузка таблиц
reviews = pd.read_csv('reviews_sample.csv')
recipes = pd.read_csv('recipes_sample.csv')

# Объединение таблиц по столбцу "id"
recipes_reviews = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id')


# Выбор числовых столбцов
numeric_cols = ['minutes', 'n_steps', 'n_ingredients', 'rating']
recipes_reviews_numeric = recipes_reviews[numeric_cols]

# Построение корреляционной матрицы
corr_matrix = recipes_reviews_numeric.corr()

# Построение тепловой карты
sns.set(style='white')
sns.set(font_scale=1.2)
heatmap = sns.heatmap(corr_matrix, annot=True, cmap='YlOrRd')
heatmap.set_title('Корреляционная матрица числовых столбцов таблиц recipes и reviews')
````
